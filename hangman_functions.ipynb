```python
def isletter(letter: str) -> bool:
  return len(letter) == 1 and letter.isalpha()
```

```python
def userinput() -> str:
  while True:
    guess = input("Enter a letter: ").lower()
    if isletter(guess):
      break
    print("Thats not a letter")
  return guess
```

```python
def crypt(word: str):
  cryptword = ""
  for l in word:
    cryptword += "_"
  return cryptword
```

```python
def replace_letter(word: str, hidden_word: str, letter: str) -> str:
  new_word = ""
  for i, l in enumerate(word.replace(" ", "")):
    new_word += letter if letter == hidden_word[i] else l
  return new_word
```

```python
def spacing(word: str) -> str:
  new_word = ""
  for l in word:
    new_word += l + " "
  return new_word.strip()
```
