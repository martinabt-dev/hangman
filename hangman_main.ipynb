```python
import hangman_functions
import hangman_words
import hangmanpics
import random
```

```python
hidden_word = random.choice(hangman_words.wordlist).lower()
word = hangman_functions.spacing(hangman_functions.crypt(hidden_word))
guessed_letters = set()
mistakes = 0
print(hangmanpics.hangmanpics[mistakes])

while mistakes < 6:

  print(word)
  guess = hangman_functions.userinput()

  if guess in guessed_letters:
    print("You already guessed this letter!")
    continue
  else:
    guessed_letters.add(guess)

  if guess in hidden_word:
    word = hangman_functions.replace_letter(word, hidden_word, guess)
    if word == hidden_word:
      print(word)
      print("You won")
      break
    word = hangman_functions.spacing(word)
  else:
    mistakes += 1
    print(hangmanpics.hangmanpics[mistakes])

if mistakes >= 6:
  print("You lose")
  print(f"The hidden word was {hidden_word}")
```
